```python
import pandas as pd
import numpy as np
from scipy.stats import median_abs_deviation
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# PAR√ÅMETROS
# ===============================
excel_path = "/content/drive/MyDrive/SGC_GGA.xlsx"

id_cols = [
    "CODIGO_MUESTRA","DEPARTAMENTO","MUNICIPIO","VEREDA",
    "Cuenca_Principal_QuebradaRio","Cuenca_Muestreada_CanoLago",
    "WGS84_LATITUD","WGS84_LONGITUD","ALTITUD",
    "CONTROL_CALIDAD","TAMIZ","Fase"
]

vars_mmad = [
    "Ag_ppm","Al_per","As_ppm","Au_ppm","B_ppm","Ba_ppm","Be_ppm","Bi_ppm",
    "C_per","Ca_per","Cd_ppm","Ce_ppm","Co_ppm","Cr_ppm","Cs_ppm","Cu_ppm",
    "Dy_ppm","Er_ppm","Eu_ppm","Fe_per","Ga_ppm","Gd_ppm","Ge_ppm","Hf_ppm",
    "Hg_ppm","Ho_ppm","In_ppm","K_per","LOI_per","La_ppm","Li_ppm","Lu_ppm",
    "Mg_per","Mn_ppm","Mo_ppm","Na_per","Nb_ppm","Nd_ppm","Ni_ppm","P_per",
    "Pb_ppm","Pd_ppm","Pr_ppm","Pt_ppm","Rb_ppm","Re_ppm","S_per","Sb_ppm",
    "Sc_ppm","Se_ppm","Si_per","Sm_ppm","Sn_ppm","Sr_ppm","Ta_ppm","Tb_ppm",
    "Te_ppm","Th_ppm","Ti_per","Tl_ppm","Tm_ppm","U_ppm","V_ppm","W_ppm",
    "Y_ppm","Yb_ppm","Zn_ppm","Zr_ppm"
]

mmad_factor = 2

# ===============================
# 1. LECTURA EXCEL
# ===============================
df = pd.read_excel(excel_path)

# ===============================
# 2. NORMALIZAR DECIMALES (, ‚Üí .)
# ===============================
for col in vars_mmad:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .astype(float)
    )

# ===============================
# 3. FORMATO LARGO
# ===============================
df_long = df[id_cols + vars_mmad].melt(
    id_vars=id_cols,
    value_vars=vars_mmad,
    var_name="Elemento",
    value_name="Valor"
)

# ===============================
# 4. TRATAMIENTO DE DL (CENSURA)
# ===============================
df_long["DL_flag"] = df_long["Valor"] < 0

# Reemplazo cl√°sico: DL = |valor| / 2
df_long.loc[df_long["DL_flag"], "Valor"] = (
    df_long.loc[df_long["DL_flag"], "Valor"].abs() / 2
)

# ===============================
# 5. LOG10 SOLO PARA VALORES V√ÅLIDOS
# ===============================
df_long = df_long[df_long["Valor"] > 0].copy()
df_long["logValor"] = np.log10(df_long["Valor"])

# ===============================
# 6. BASELINE mMAD
# ===============================
baseline = []

for elem, g in df_long.groupby("Elemento"):
    x = g["logValor"].dropna()

    med = np.median(x)
    mmad = median_abs_deviation(x, scale="normal")

    baseline.append({
        "Elemento": elem,
        "Median_log": med,
        "mMAD": mmad,
        "Upper": med + mmad_factor * mmad,
        "Lower": med - mmad_factor * mmad
    })

baseline = pd.DataFrame(baseline)

# ===============================
# 7. OUTLIERS
# ===============================
df_long = df_long.merge(baseline, on="Elemento", how="left")

df_long["Outlier_mMAD"] = (
    (df_long["logValor"] > df_long["Upper"]) |
    (df_long["logValor"] < df_long["Lower"])
)
# ===============================
# 8. GR√ÅFICOS INDIVIDUALES (UNO POR ELEMENTO)
#    CON BARRA mMAD Y S√çMBOLOS TIPO FIGURA
# ===============================

for elem in baseline["Elemento"]:

    df_e = df_long[df_long["Elemento"] == elem]
    base_e = baseline[baseline["Elemento"] == elem].iloc[0]

    plt.figure(figsize=(4, 6))

    # -------------------------------------------------
    # BARRA DE COLOR DE FONDO (VERDE ‚Üí AMARILLO)
    # -------------------------------------------------
    palette_6 = [
        "#0b3d0b",  # verde muy oscuro
        "#1f7a1f",
        "#4caf50",
        "#9ccc65",
        "#dce775",
        "#fff59d"   # amarillo claro
    ]

    med = base_e["Median_log"]
    mmad = base_e["mMAD"]

    bounds = [
        med - 3*mmad,
        med - 2*mmad,
        med - 1*mmad,
        med,
        med + 1*mmad,
        med + 2*mmad,
        med + 3*mmad
    ]

    for i in range(6):
        plt.axhspan(
            bounds[i],
            bounds[i+1],
            color=palette_6[i],
            alpha=0.40,
            zorder=0
        )

    # -------------------------------------------------
    # BOXPLOT CON S√çMBOLOS PERSONALIZADOS
    # -------------------------------------------------
    sns.boxplot(
        y=df_e["logValor"],
        width=0.28,
        showfliers=False,
        showmeans=True,                         # üîß s√≠mbolo estad√≠stico
        meanprops=dict(
            marker="o",                         # ‚óè negro como en la figura
            markerfacecolor="black",
            markeredgecolor="black",
            markersize=5
        ),
        boxprops=dict(
            facecolor="none",
            edgecolor="black",
            linewidth=1.2
        ),
        whiskerprops=dict(
            color="black",
            linewidth=1.4
        ),
        capprops=dict(
            color="black",
            linewidth=1.4
        ),
        medianprops=dict(
            color="black",
            linewidth=0.8,                      # mediana no dominante
            linestyle="-"
        ),
        zorder=2
    )

    # -------------------------------------------------
    # PUNTOS INDIVIDUALES (JITTER)
    # -------------------------------------------------
    sns.stripplot(
        y=df_e["logValor"],
        color="black",
        size=3,
        alpha=0.65,
        jitter=0.12,
        zorder=3
    )

    # -------------------------------------------------
    # L√çNEAS mMAD (REFERENCIA)
    # -------------------------------------------------
    plt.axhline(med, color="black", linewidth=2.0)
    plt.axhline(med + 2*mmad, color="black", linestyle="--", linewidth=1.4)
    plt.axhline(med - 2*mmad, color="black", linestyle="--", linewidth=1.4)

    # -------------------------------------------------
    # EST√âTICA FINAL
    # -------------------------------------------------
    plt.ylabel("log10(concentration)")
    plt.title(f"{elem} ‚Äî mMAD baseline", fontsize=11)
    plt.grid(axis="y", linestyle=":", alpha=0.4)

    plt.tight_layout()
    plt.show()

# ===============================
# 9. EXPORTACI√ìN
# ===============================
baseline.to_excel("mMAD_Baseline.xlsx", index=False)
df_long[df_long["Outlier_mMAD"]].to_excel("mMAD_Outliers.xlsx", index=False)
df_long.to_excel("mMAD_AllData.xlsx", index=False)

print("‚úÖ Script ejecutado correctamente con datos reales")
```
